```python
import pandas as pd
import numpy as np
np.random.seed(1234)
```

```python
def generate_trade_db(nsecs=1000, ndays=40, nqty=10, sprice=5, ntrans=100000):
    """ 
    generate a frame of trade db
    
    sample for ndays
    each day create between 0.25 and 1.0 ntrades per day
    allocated randomly among nsecs, each having a nqtys
    sprice is the price dispersion per security
    """
    
    # securities
    secs = np.array([ "S%04d" % i for i in range(nsecs)])
    
    # base price
    base_prices = Series(np.random.randn(len(secs))*15+50,index=secs)

    # directions (buy/sell)
    dirs = np.array(['buy','sell'])

    # quantity
    qtys = np.random.randint(1,1000,size=nqty)
    
    # date
    dates = pd.date_range('20130101',periods=ndays,freq='D')

    # transactions per day
    trans = np.random.randint(int(0.25*ntrans),int(1.0*ntrans),size=len(dates))

    # selection matrix
    # secs * dirs * qtys
    selections = pd.MultiIndex.from_product([secs,dirs,qtys])

    # return an index of the choices
    def choose_trades(x):
        return pd.MultiIndex.from_tuples(np.random.choice(selections.values,size=x,replace=True),
                                         names=['sec','dir','qty'])

    def create_per_day(date,trans):
        # create a frame with the taken values (trans) per day
        t = DataFrame(date,index=choose_trades(trans),columns=['date']).reset_index()
        t['price'] = t['sec'].map(base_prices)
        t['tprice'] = t.groupby('sec')['price'].apply(lambda x: np.random.randn(len(x))*10+x)
        return t

    trades = pd.concat([ create_per_day(date,trans) for date,trans in zip(dates,trans)],ignore_index=True)
    trades['dir'] = trades['dir'].astype('category')
    return trades
```

```python
trades = generate_trade_db()
trades.info()
```

```python
# number of totals trades
print "total trades"
print len(trades)

# total buys/sells
print "\ntotal number by action"
print trades.groupby('dir')['dir'].count()
print "\ntotal qty by action"
print trades.groupby('dir')['qty'].sum()

# signed qty for each trade
trades['sqty'] = trades['qty'].where(trades['dir'] == 'buy',-trades['qty'])
trades.pop('dir')

# per day, groupby by security and direction, create net, gross trades
g = trades.groupby([pd.Grouper(key='date',freq='D'),'sec'])

# net and gross traded per day
print g[['qty','sqty']].sum()

# net and gross number of trades per day
print g[['qty','sqty']].count()
```
